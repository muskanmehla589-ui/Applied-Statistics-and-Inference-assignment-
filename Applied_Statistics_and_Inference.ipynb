**Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?**


When we do hypothesis testing, we check:

-**Null Hypothesis (H₀):** Nothing is happening / no change.

-**Alternative Hypothesis (H₁):** Something is happening / there is a change.

While making decisions from data, two types of mistakes can happen.


**1. Type I Error (False Positive)**

**Meaning:**
You think something has changed, but actually nothing has changed.


**Example:**
Your A/B test shows a new button increases clicks, but actually the clicks increased just by chance.

**Impact:**

- You make a wrong decision.

- You may spend money on a change that doesn't really work.

**2. Type II Error (False Negative)**

**Meaning:**
You think nothing has changed, but actually something has changed.

**Example:**
A new marketing idea actually increases sales, but your test fails to detect it.

**Impact:**

- You miss a good opportunity.

- The business loses potential growth.


**Quick Comparison Table**

Error	Simple Meaning	Example

- Type I Error	False alarm	You think a new feature works when it doesn’t

- Type II Error	Missed detection You miss a feature that actually works



**Question 2: What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?**

A P-value tells us how likely it is to get our sample results if the null hypothesis (H₀) is actually true.

**In simple words:**

P-value = Probability that your data happened just by chance if there is really no effect.

**How to Interpret the P-value?**

1. **If P-value is small (usually < 0.05)**

The result is unlikely to be due to chance.

We reject the null hypothesis.

Meaning: Something is really happening / there is an effect.

**Example:**
P-value = 0.03
- → Only 3% chance the result came by luck.
- → So, we reject H₀.

2. **If P-value is large (usually > 0.05)**

- The result could easily be due to chance.

- We fail to reject the null hypothesis.

- Meaning: No strong evidence of change or effect.

**Example:**
P-value = 0.40
- → 40% chance the result is due to luck.
- → So, H₀ remains true.



**Question 3:Explain the difference between a Z-test and a T-test, including when to use each.**


**1. Z-test**

**When to use:**

- When the sample size is large (n > 30)

- When the population standard deviation (σ) is known

- Data follows a normal distribution

**Where used in Data Analytics:**

- A/B testing with large traffic

- Quality control

- Large-scale surveys

**Example**:
Testing if the average height of students in a college differs from a national average when the population standard deviation is known.

**2. T-test**

**When to use:**

- When the sample size is small (n < 30)

- When the population standard deviation is NOT known

- Data is approximately normal or unknown

**Where used in Data Analytics:**

- Comparing means in small experiments

- Small A/B test samples

- Checking if a product change affects user behavior with limited data

**Example:**
Testing if the average marks of 20 students differ from 50 when σ is unknown.


**In Simple Words**:

- Use Z-test when you have lots of data and know the population standard deviation.

- Use T-test when you have small data and do not know the population standard deviation.

**Question 4: What is a confidence interval, and how does the margin of error influence its width and interpretation?**


A confidence interval (CI) is a range of values that is likely to contain the true population value.

**It usually looks like:**

- Estimate ± Margin of Error

Example: 50 ± 5 → Confidence Interval = 45 to 55

**Meaning:**
We are confident (often 95% confident) that the true value lies somewhere between 45 and 55.

**What is Margin of Error?**

The margin of error tells us how much the estimate could vary due to random sampling.

**How Margin of Error Affects the Width of the CI?**

1. **Larger Margin of Error** → Wider Confidence Interval

- More uncertainty

- Less precise estimate

- Example: 50 ± 10 → CI = 40 to 60 (very wide)

2. **Smaller Margin of Error** → Narrower Confidence Interval

- Less uncertainty

- More precise estimate

- Example: 50 ± 2 → CI = 48 to 52 (narrow)

**What influences the Margin of Error?**

**Sample size:**
Larger sample → smaller margin of error → narrower CI

**Variability in data:**
More variability → larger margin of error

**Confidence level:**
Higher confidence (99%) → bigger margin of error than 95%

**Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?**

**ANOVA (Analysis of Variance)**:  ANOVA is used to compare the means of three or more groups to check if they are significantly different.

**Simple meaning:**
 ANOVA tells us whether all groups have the same average or at least one group is different.

**Example:**
Comparing average marks of students from 3 different classes.

**Why Use ANOVA Instead of Multiple T-tests?**

If we compare many groups using separate t-tests, the chance of error increases.

- ANOVA avoids this problem and tests all groups at once.

**Assumptions of ANOVA**

For ANOVA results to be valid, these conditions must be met:

1. **Independence of observations**

Each data point should be collected independently.

2. **Normality**

The data in each group should be approximately normally distributed.

3. **Homogeneity of variances**

All groups should have similar variances (similar spread).

**How ANOVA Extends Hypothesis Testing Beyond Two Groups**

Normally, t-tests compare only two groups.
 ANOVA extends this by comparing three or more means at the same time.

**Hypotheses in ANOVA:**

**Null Hypothesis (H₀):**
All group means are equal
(No difference between groups)

**Alternative Hypothesis (H₁):**
At least one group mean is different

ANOVA does not tell which group is different; it only says a difference exists.
If ANOVA is significant, we use post-hoc tests (like Tukey test) to find the exact groups that differ.

**Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.**

```python
import numpy as np
from math import sqrt
from scipy.stats import norm

# Sample dataset
data = [52, 55, 48, 50, 49, 53, 54, 51, 50, 52]

# Given information
population_mean = 50        # μ0
population_std = 5          # σ (known population standard deviation)
n = len(data)

# Step 1: Calculate sample mean
sample_mean = np.mean(data)

# Step 2: Calculate Z-statistic
z_value = (sample_mean - population_mean) / (population_std / sqrt(n))

# Step 3: Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_value)))

# Print results
print("Sample Mean:", sample_mean)
print("Z-value:", z_value)
print("P-value:", p_value)

# Step 4: Interpretation
alpha = 0.05  # significance level

if p_value < alpha:
    print("Result: Reject the null hypothesis (Significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (No significant difference).")
```

Explanation (Simple)

We calculate the sample mean.

**Use the Z-test formula:**


Compute the p-value.

Compare p-value with α = 0.05 to decide if the result is significant.

**Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.**


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a non-normal distribution (Exponential)
data = np.random.exponential(scale=2, size=10000)

# Step 2: Apply Central Limit Theorem
sample_means = []
sample_size = 50   # sample size for CLT

for _ in range(1000):  # number of samples
    sample = np.random.choice(data, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Step 3: Plot original non-normal distribution
plt.hist(data, bins=40)
plt.title("Original Non-Normal Distribution (Exponential)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 4: Plot distribution of sample means (CLT Result)
plt.hist(sample_means, bins=30)
plt.title("Sampling Distribution of the Mean (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

Your Central Limit Theorem simulation is complete!
- The first plot shows a non-normal exponential distribution (skewed right).
- The second plot shows that the sampling distribution of the means becomes approximately normal, even though the original data is not — this demonstrates the Central Limit Theorem.

**Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_confidence_interval(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for the sample mean.
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    # Z-score for the given confidence level
    z = norm.ppf((1 + confidence) / 2)

    margin_of_error = z * (sample_std / np.sqrt(n))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Plot
    plt.figure(figsize=(8,4))
    plt.axvline(sample_mean)
    plt.axvline(lower_bound)
    plt.axvline(upper_bound)
    plt.title("Confidence Interval for Sample Mean")
    plt.xlabel("Value")
    plt.ylabel("Reference Lines")
    plt.show()

    return sample_mean, (lower_bound, upper_bound)

# Example dataset
data = np.random.normal(loc=50, scale=10, size=100)

plot_confidence_interval(data)
```

 - **What the function does**

- Takes a dataset

- **Calculates:**

1.   Sample mean
1.   Standard deviation
2.   Margin of error
2.   Confidence interval


- **Draws a plot with:**

1.   A line for the sample mean
2.   Lines for the lower & upper confidence interval bounds


**Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.**


```python
import numpy as np
from scipy.stats import chisquare

observed = [25, 30, 20, 25]
expected = [25, 25, 25, 25]

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)
```

This Python code performs a Chi-square goodness-of-fit test by comparing the observed frequencies with the expected frequencies. The **chisquare()** function calculates two things: the Chi-square statistic and the p-value. The Chi-square statistic shows how much the observed values differ from what was expected, and the p-value tells us whether this difference is statistically significant. If the p-value is greater than 0.05, we conclude that there is no significant difference between the observed and expected distributions, meaning the data fits the expected pattern well. If the p-value is less than 0.05, then the difference is considered significant, meaning the observed distribution does not fit the expected distribution.
